```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data_enc = pd.read_csv('Poll_answers_encoded.csv')
```

#### Visualization

```python
cols = ['freq_coding','git_repo','git_branch','unit_test','linters','soft_doc','debuggers','git_PRs','collab_coding','paradigms','SDDs']
```

```python
plt.scatter(data_enc['stage'],data_enc['training'],alpha=0.3)
```

```python
fig, axs = plt.subplots(nrows=len(cols)//3+1,ncols=3,figsize=(15,15),sharey='row')
for i,col in enumerate(cols):
    ax = axs.flatten()[i]
    ax.scatter(data_enc['stage'],data_enc[col],alpha=0.2,s=80)
    ax.tick_params(axis='x', labelrotation=45)
    ax.set_title(col)
axs.flatten()[-1].set_visible(False)
plt.tight_layout()
```

```python
fig, axs = plt.subplots(nrows=len(cols)//3+1,ncols=3,figsize=(10,10),sharey='row')
for i,col in enumerate(cols):
    ax = axs.flatten()[i]
    values = data_enc[col].value_counts().values
    labels = pd.Series(data_enc[col].value_counts().index)
    labels = list(labels.replace({1:'never',2:'rarely',3:'sometimes',4:'always'}))
    patches,_ = ax.pie(values,labels=labels)
    for p in patches:
        if p.get_label()=='always': 
            p.set_facecolor('#eb5334')
        if p.get_label()=='sometimes': 
            p.set_facecolor('#ebb434')
        if p.get_label()=='rarely': 
            p.set_facecolor('#3aba71') 
        if p.get_label()=='never': 
            p.set_facecolor('#54e3de') 
    ax.set_title(col)
axs.flatten()[-1].set_visible(False)
plt.show()
plt.tight_layout();
```
